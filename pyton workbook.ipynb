```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Name": ["A","B","C","D"],
    "Marks": [90,70,60,77]
}
df = pd.DataFrame(data)
df.plot(x="Name", y="Marks", kind="bar")
plt.show()
```

```python
import pandas as pd
data = {"Name": ["A", "B", "C"], "Marks": [90, 40, 75]}
df = pd.DataFrame(data)
df["Result"] = df["Marks"].apply(lambda x : "Pass" if x>50 else "Fail")
print(df)
```
