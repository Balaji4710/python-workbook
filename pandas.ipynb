```python
import pandas as pd
import numpy as np
data = {
    "Name": ['A','B','C','D','E'],
    "Department": ["IT","HR","IT","Finance","HR"],
    "Salary" : [50000,40000,60000,55000,45000],
    "Experience": [2,3,5,4,2]
    
}
df = pd.DataFrame(data)
print(df)
print("Highest Salary")
print(df[df['Salary']== df['Salary'].max()])
print("Average Salary by Department")
print(df.groupby("Department")["Salary"].mean())
df["Bonus"] = df["Salary"]*0.10
print("Experienced Employees")
print(df[df["Experience"]>3])
print("Updated Data\n", df)
```

```python
import numpy as np
import pandas as pd
np.random.seed(0)
products = ["Laptop", "Mobile", "Tablet", "Headphone"]
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales_data = np.random.randint(1000, 5000, size=(5, 4))
df = pd.DataFrame(sales_data, columns=products, index=months)
print("Sales Data:\n", df)
print("\nTotal Sales per Product:")
print(df.sum())
print("\nHighest Sales Month:")
print(df.sum(axis=1).idxmax())
print("\nBest Selling Product:")
print(df.sum().idxmax())
print("\nAverage Sales:")
print(df.mean())
import matplotlib.pyplot as plt
df.plot(kind='bar')
plt.title("Monthly Sales")
plt.show()
```

```python
import numpy as np
import pandas as pd
data = {
    "StudyHours": [1, 2, 3, 4, 5, 6, 7],
    "Marks": [30, 40, 50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)
print("Correlation:\n", df.corr())
def predict_marks(hours):
    return 10 * hours + 20
df["PredictedMarks"] = df["StudyHours"].apply(predict_marks)
print("\nPredicted Data:\n", df)
```

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["A", "B", None, "D"],
    "Age": [20, np.nan, 22, 21],
    "Marks": [80, 90, None, 70]
}
df = pd.DataFrame(data)
print("Before cleaning",df)
df["Name"].fillna("Unknown", inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(0, inplace=True)
df.drop_duplicates(inplace =True)
print("After cleaning\n",df)
```
