```python
def run_quiz():
    questions = [
        {
            "question": "What is the capital of France?",
            "options": ["A) Paris", "B) London", "C) Rome", "D) Berlin"],
            "answer": "A"
        },
        {
            "question": "Which programming language is known as the 'snake' language?",
            "options": ["A) Java", "B) Python", "C) C++", "D) Ruby"],
            "answer": "B"
        },
        {
            "question": "What is 5 + 7?",
            "options": ["A) 10", "B) 11", "C) 12", "D) 13"],
            "answer": "C"
        },
        {
            "question" : "The Aryans dedicated whole of a separate Mandala in Rig-Veda for the hymns dedicated to __?",
            "options": ["A) Agni", "B) Varuna", "C) Indra","D) Soma"],
            "answer": "D"
        },
        {
            "question": "What is the IUCN conservation status of “Mugger Crocodile” that was recently seen in news?",
            "options": ["A) Critically Endangered", "B) Endangered", "C) Vulnerable ","D) Least Concern"],
            "answer": "C"
        },
{
            "question": "Which state/UT is home to the highest recorded density of snow leopards in the world?",
            "options": ["A) Uttarakand", "B) Ladakh ", "C) Mizoram", "D) Sikkim "],
            "answer": "B"
},
        {
            "question": "What is the currency of India?",
            "options": ["A) Rupees", "B) Dollar", "C) Dinhams", "D) Yen"],
            "answer": "A"
        },
        {
            "question": "What is the first Indian to land on the moon?",
            "options": ["A) Kalpana Chawala", "B) Neil Amstrong", "C) Yuri gargarin", "D) Rakesh Sharma"],
            "answer": "D"
        },
        {
            "question": "What is the capital of South Korea?",
            "options": ["A) Paris", "B) Seoul", "C) Rome", "D) Berlin"],
            "answer": "B"
        },
        {
            "question": "What is the capital of United Kingdom?",
            "options": ["A) Paris", "B) London", "C) Rome", "D) Berlin"],
            "answer": "B"
        }
    ]

    score = 0

    print("--- Welcome to the Quiz! ---")
    print("Please answer by typing the letter (A, B, C, or D).\n")

    for i, q in enumerate(questions):
        print(f"Question {i + 1}: {q['question']}")
        for option in q['options']:
            print(option)
    
        guess = input("Your answer: ").strip().upper()
        if guess == q['answer']:
            print("Correct!\n")
            score += 1
        else:
            print(f"Wrong! The correct answer is {q['answer']}.\n")

    total_questions = len(questions)
    percentage = (score / total_questions) * 100
    
    print("--- Quiz Results ---")
    print(f"Final Score: {score}/{total_questions}")
    print(f"Percentage: {percentage:.1f}%")
run_quiz()
```
