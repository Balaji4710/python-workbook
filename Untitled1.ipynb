```python
def count_frequency(lst):
    freq_dict = {}
    for elem in lst:
        if elem in freq_dict:
            freq_dict[elem] += 1
        else:
            freq_dict[elem] = 1
    return freq_dict

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
freq_dict = count_frequency(numbers)
print(freq_dict)
```

```python
def merge_dicts(dict1, dict2):
    return {**dict1, **dict2}

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged_dict = merge_dicts(dict1, dict2)
print(merged_dict)
```

```python
def sort_dict_by_keys(d):
    return dict(sorted(d.items()))

my_dict = {'c': 3, 'a': 1, 'b': 2}
sorted_dict = sort_dict_by_keys(my_dict)
print(sorted_dict)
```

```python
def sort_dict_by_values(d):
    return dict(sorted(d.items(), key=lambda item: item[1]))

my_dict = {'a': 3, 'b': 1, 'c': 2}
sorted_dict = sort_dict_by_values(my_dict)
print(sorted_dict)
```

```python
def max_key(d):
    return max(d, key=d.get)

my_dict = {'a': 10, 'b': 20, 'c': 30}
max_key_val = max_key(my_dict)
print(max_key_val)
```

```python
def key_exists(d, key):
    return key in d

my_dict = {'a': 1, 'b': 2, 'c': 3}
exists = key_exists(my_dict, 'b')
print(exists)
```
