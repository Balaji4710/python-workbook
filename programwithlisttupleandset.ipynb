```python
orders = ["apple", "banana", "apple", "mango", "banana", "grape", "mango", "apple"]
unique_fruits = set(orders)
print(unique_fruits)
pouplar_fruits = []
fruits = []
for i in orders:
    if (orders.count(i)>1):
        pouplar_fruits.append(i)
print(pouplar_fruits)
for i in pouplar_fruits:
    if i not in fruits:
        fruits.append(i)
print(fruits)

```

```python
students = [
    ("Arun", [85, 90, 78]),
    ("Bala", [88, 76, 90]),
    ("Charan", [85, 90, 78]),
    ("Divya", [95, 92, 88])
]

for student in students:
    name = student[0]        
    marks = student[1]  
    average = sum(marks) / len(marks)  
    print(f"{name}'s average: {average:.2f}")
```

```python
records = [
    (101, "Math"),
    (102, "Science"),
    (101, "Math"),
    (103, "English"),
    (102, "Science"),
    (101, "English"),
    (104, "Math")
]
id = set()
subject = []
for i in records:
    id.add(i[0])
    subject.append(i[1])
print(id)
res = []
for i in subject: 
    if i not in res:
        res.append(i)
print(res)
```

```python
raw_sales = [
    (101, "General"), (102, "VIP"), (101, "General"), 
    (103, "General"), (102, "VIP"), (104, "Backstage"),
    (105, "VIP"), (101, "General"), (106, "General")
]
customer_id= set()
passType = []
for i in raw_sales:
    customer_id.add(i[0])
    passType.append(i[1])
permuim_pass=[]
for i in passType:
    if (i != "General"):
        permuim_pass.append(i)
print(permuim_pass)
```
