```python
from abc import ABC,abstractmethod
class Car(ABC):
    def mileage(self):
        pass
class Tesla(Car):
    def mileage(self):
        print("The mileage is 30kmph")
class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25kmph")
class Duster(Car):
    def mileage(self):
        print("The mileage is 24kmph ")
class Renault(Car):
    def mileage(self):
        print("The mileage is 37kmph ")
t=Tesla()
t.mileage()
r=Renault()
r.mileage()

s=Suzuki()
s.mileage()
d=Duster()
d.mileage()
```

```python
class Birds:
    def intro1(self):
        print("There are multiple types of birds in the world.")
    def flight1(self):
        print("Many of these birds can fly but some cannot.")
class sparrow1(Birds):
    def flight1(self):
        print("Sparrows are the bird which can fly.")
class ostrich1(Birds):
    def flight1(self):
        print("Ostriches are the birds which cannot fly.")
obj_birds=Birds()
obj_spr1=sparrow1()
obj_ost1=ostrich1()
obj_birds.intro1()
obj_birds.flight1()
obj_spr1.intro1()
obj_spr1.flight1()
obj_ost1.intro1()
obj_ost1.flight1()
```

```python
import threading
import time
def cal_square(num):
    print("Calculating square of number")
    for n in num:
        time.sleep(0.3)
        print("square is ",n*n)
arr = [4,5,6,7,8]
t1=time.time()
cal_square(arr)
print("total time taken by threads is ",time.time()-t1)
```
