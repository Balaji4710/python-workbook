```python
students = [
    ["Arun", [80, 75, 90]],
    ["Bala", [60, 70, 65]],
    ["Cathy", [95, 85, 92]]
    
]
topper =""
maxavg = 0
for i in students:
    name = i[0]
    marks = i[1]
    average = sum(marks)/len(marks)
    print(f"The student's Name is {name} and average is {average}")
    if average >maxavg:
        maxavg = average
        topper = name
print("the topper is ", topper)
```

```python
students = [
    {"name": "Arun", "course": "AI"},
    {"name": "Bala", "course": "ML"},
    {"name": "Arun", "course": "AI"}
]

unique_students = []
seen = set()

for s in students:
    identifier = (s["name"], s["course"])
    print(identifier)

    if identifier not in seen:
        seen.add(identifier)
        unique_students.append(s)

print("Unique Students:")
for u in unique_students:
    print(u)
```

```python
import pandas as pd
import numpy as np
data = {
    "Salary": [20000, 30000, 40000, 1000000, 35000, 45000],}
df = pd.DataFrame(data)
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1
filtered_df = df[(df["Salary"] >= Q1 - 1.5*IQR) & (df["Salary"] <= Q3 + 1.5*IQR)]
print("Before:\n", df)
print("After:\n", filtered_df)
```

```python
import numpy as np
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
np.random.seed(1)
weights = np.random.rand(2,1)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
for i in range(10000):
    input_layer = X
    output = sigmoid(np.dot(input_layer, weights))
    error = y - output
    adjustments = error * (output * (1 - output))
    weights += np.dot(input_layer.T, adjustments)
print("Predictions:\n", output)
```

```python
import pandas as pd
import numpy as np
dates = pd.date_range(start="2024-01-01", periods=100)
sales = np.random.randint(1000, 5000, 100)
df = pd.DataFrame({"Date": dates, "Sales": sales})
df.set_index("Date", inplace=True)
df["RollingMean"] = df["Sales"].rolling(window=5).mean()
df["Trend"] = df["Sales"].diff()
print(df.tail())
```
